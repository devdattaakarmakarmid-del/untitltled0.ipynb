```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import requests
```

```python
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"

text = requests.get(url).text

print("Total Characters :", len(text))
print()
print(text[:500])
```

```python
characters = sorted(list(set(text)))

vocab_size = len(characters)

print("Vocabulary Size :", vocab_size)

# Character to Integer
char_to_int = {}

# Integer to Character
int_to_char = {}

for i, ch in enumerate(characters):
    char_to_int[ch] = i
    int_to_char[i] = ch
```

```python
def encode(sentence):

    numbers = []

    for ch in sentence:
        numbers.append(char_to_int[ch])

    return numbers


def decode(numbers):

    sentence = ""

    for num in numbers:
        sentence += int_to_char[num]

    return sentence
```

```python
class ShakespeareDataset(Dataset):

    def __init__(self, data, sequence_length):

        self.data = data
        self.sequence_length = sequence_length

    def __len__(self):

        return len(self.data) - self.sequence_length

    def __getitem__(self, index):

        x = self.data[index:index+self.sequence_length]

        y = self.data[index+1:index+self.sequence_length+1]

        return x, y
```

```python
##################################################
# Multi Head Attention
##################################################

import math
import torch
import torch.nn as nn


class MultiHeadAttention(nn.Module):

    def __init__(self, d_model, num_heads):

        super().__init__()

        self.d_model = d_model
        self.num_heads = num_heads

        self.head_dim = d_model // num_heads

        assert self.head_dim * num_heads == d_model

        #####################################

        self.query = nn.Linear(
            d_model,
            d_model
        )

        self.key = nn.Linear(
            d_model,
            d_model
        )

        self.value = nn.Linear(
            d_model,
            d_model
        )

        self.fc_out = nn.Linear(
            d_model,
            d_model
        )

        self.dropout = nn.Dropout(
            0.1
        )

    def forward(self, x):

        batch_size = x.shape[0]
        seq_len = x.shape[1]

        ####################################
        # Query Key Value
        ####################################

        Q = self.query(x)
        K = self.key(x)
        V = self.value(x)

        ####################################
        # Split Heads
        ####################################

        Q = Q.view(
            batch_size,
            seq_len,
            self.num_heads,
            self.head_dim
        ).transpose(1,2)

        K = K.view(
            batch_size,
            seq_len,
            self.num_heads,
            self.head_dim
        ).transpose(1,2)

        V = V.view(
            batch_size,
            seq_len,
            self.num_heads,
            self.head_dim
        ).transpose(1,2)

        ####################################
        # Attention Score
        ####################################

        score = torch.matmul(

            Q,

            K.transpose(-2,-1)

        )

        score = score / math.sqrt(

            self.head_dim

        )

        ####################################
        # Causal Mask
        ####################################

        mask = torch.tril(

            torch.ones(

                seq_len,

                seq_len,

                device=x.device

            )

        )

        score = score.masked_fill(

            mask == 0,

            float("-inf")

        )

        ####################################
        # Softmax
        ####################################

        attention = torch.softmax(

            score,

            dim=-1

        )

        attention = self.dropout(

            attention

        )

        ####################################
        # Weighted Sum
        ####################################

        output = torch.matmul(

            attention,

            V

        )

        ####################################
        # Merge Heads
        ####################################

        output = output.transpose(

            1,

            2

        ).contiguous()

        output = output.view(

            batch_size,

            seq_len,

            self.d_model

        )

        ####################################
        # Final Linear Layer
        ####################################

        output = self.fc_out(

            output

        )

        return output
```

```python
class FeedForward(nn.Module):

    def __init__(self, d_model):

        super().__init__()

        self.network = nn.Sequential(

            nn.Linear(d_model, d_model * 4),

            nn.GELU(),

            nn.Dropout(0.1),

            nn.Linear(d_model * 4, d_model)

        )

    def forward(self, x):

        return self.network(x)
```

```python
##################################################
# Residual Connection
##################################################

class ResidualConnection(nn.Module):

    def __init__(self):

        super().__init__()

    def forward(self, x, sublayer_output):

        return x + sublayer_output
```

```python
##################################################
# Layer Normalization
##################################################

class LayerNormalization(nn.Module):

    def __init__(self, embedding_size):

        super().__init__()

        self.layer_norm = nn.LayerNorm(
            embedding_size
        )

    def forward(self, x):

        return self.layer_norm(x)
```

```python
##################################################
# Transformer Block
##################################################

class TransformerBlock(nn.Module):

    def __init__(
        self,
        d_model,
        num_heads
    ):
        super().__init__()

        ##################################
        # Multi Head Attention
        ##################################

        self.attention = MultiHeadAttention(
            d_model,
            num_heads
        )

        ##################################
        # Feed Forward
        ##################################

        self.feed_forward = FeedForward(
            d_model
        )

        ##################################
        # Residual Connections
        ##################################

        self.residual1 = ResidualConnection()
        self.residual2 = ResidualConnection()

        ##################################
        # Layer Normalization
        ##################################

        self.layer_norm1 = LayerNormalization(
            d_model
        )
        self.layer_norm2 = LayerNormalization(
            d_model
        )

    def forward(self, x):

        ##################################
        # Multi Head Attention
        ##################################

        attention_output = self.attention(x)

        ##################################
        # Residual Connection 1
        ##################################

        x = self.residual1(
            x,
            attention_output
        )

        ##################################
        # Layer Normalization 1
        ##################################

        x = self.layer_norm1(x)

        ##################################
        # Feed Forward Network
        ##################################

        feed_forward_output = self.feed_forward(
            x
        )

        ##################################
        # Residual Connection 2
        ##################################

        x = self.residual2(
            x,
            feed_forward_output
        )

        ##################################
        # Layer Normalization 2
        ##################################

        x = self.layer_norm2(x)

        return x
```

```python
class SimpleTransformer(nn.Module):

    def __init__(self,
                 vocab_size,
                 d_model,
                 num_heads,
                 max_seq_length):

        super().__init__()

        self.dropout = nn.Dropout(0.1)

        self.token_embedding = nn.Embedding(vocab_size, d_model)

        self.position_embedding = nn.Embedding(max_seq_length, d_model)

        self.transformer = TransformerBlock(
            d_model,
            num_heads
        )

        self.output_layer = nn.Linear(
            d_model,
            vocab_size
        )

    def forward(self, x):

        batch_size = x.size(0)
        seq_len = x.size(1)

        positions = torch.arange(seq_len, device=x.device)
        positions = positions.unsqueeze(0).expand(batch_size, seq_len)

        token = self.token_embedding(x)
        position = self.position_embedding(positions)

        x = token + position
        x = self.dropout(x)

        x = self.transformer(x)

        logits = self.output_layer(x)

        return logits
```

```python
import torch.nn as nn

# Redefine TransformerBlock to match SimpleTransformer's expected constructor
class TransformerBlock(nn.Module):

    def __init__(
        self,
        d_model,
        num_heads
    ):
        super().__init__()

        ##################################
        # Multi Head Attention
        ##################################

        self.attention = MultiHeadAttention(
            d_model,
            num_heads
        )

        ##################################
        # Feed Forward
        ##################################

        self.feed_forward = FeedForward(
            d_model
        )

        ##################################
        # Residual Connections
        ##################################

        self.residual1 = ResidualConnection()
        self.residual2 = ResidualConnection()

        ##################################
        # Layer Normalization
        ##################################

        self.layer_norm1 = LayerNormalization(
            d_model
        )
        self.layer_norm2 = LayerNormalization(
            d_model
        )

    def forward(self, x):

        ##################################
        # Multi Head Attention
        ##################################

        attention_output = self.attention(x)

        ##################################
        # Residual Connection 1
        ##################################

        x = self.residual1(
            x,
            attention_output
        )

        ##################################
        # Layer Normalization 1
        ##################################

        x = self.layer_norm1(x)

        ##################################
        # Feed Forward Network
        ##################################

        feed_forward_output = self.feed_forward(
            x
        )

        ##################################
        # Residual Connection 2
        ##################################

        x = self.residual2(
            x,
            feed_forward_output
        )

        ##################################
        # Layer Normalization 2
        ##################################

        x = self.layer_norm2(x)

        return x

model = SimpleTransformer(
    vocab_size=vocab_size,
    d_model=128,
    num_heads=8,
    max_seq_length=64
)

print(model)
```

```python
def generate_text(
        model,
        start_text,
        max_new_tokens=200
):

    model.eval()

    encoded = encode(start_text)

    input_ids = torch.tensor(
        encoded,
        dtype=torch.long
    ).unsqueeze(0).to(device)

    generated = encoded.copy()

    with torch.no_grad():

        for _ in range(max_new_tokens):

            current_input = input_ids[:, -max_seq_length:]

            output = model(current_input)

            next_token = torch.argmax(
                output[:, -1, :],
                dim=-1
            )

            generated.append(next_token.item())

            input_ids = torch.cat(
                [
                    input_ids,
                    next_token.unsqueeze(0)
                ],
                dim=1
            )

    return decode(generated)
```

```python
def generate_temperature(

        model,

        start_text,

        temperature=1.0,

        max_new_tokens=200

):

    model.eval()

    encoded = encode(start_text)

    input_ids = torch.tensor(
        encoded,
        dtype=torch.long
    ).unsqueeze(0).to(device)

    generated = encoded.copy()

    with torch.no_grad():

        for _ in range(max_new_tokens):

            current_input = input_ids[:, -max_seq_length:]

            output = model(current_input)

            logits = output[:, -1, :]

            logits = logits / temperature

            probabilities = torch.softmax(
                logits,
                dim=-1
            )

            next_token = torch.multinomial(
                probabilities,
                num_samples=1
            )

            generated.append(next_token.item())

            input_ids = torch.cat(
                [
                    input_ids,
                    next_token
                ],
                dim=1
            )

    return decode(generated)
```

```python
def generate_top_k(
        model,
        start_text,
        max_new_tokens=200,
        temperature=1.0,
        k=20
):

    model.eval()

    # Encode starting text
    encoded = encode(start_text)

    input_ids = torch.tensor(
        encoded,
        dtype=torch.long
    ).unsqueeze(0).to(device)

    generated = encoded.copy()

    with torch.no_grad():

        for _ in range(max_new_tokens):

            # Keep only last max_seq_length tokens
            current_input = input_ids[:, -max_seq_length:]

            # Forward pass
            output = model(current_input)

            # Get last token logits
            logits = output[:, -1, :]

            # Apply temperature
            logits = logits / temperature

            # Get Top-k values
            top_k_values, top_k_indices = torch.topk(
                logits,
                k
            )

            # Convert to probabilities
            probabilities = torch.softmax(
                top_k_values,
                dim=-1
            )

            # Randomly choose one token
            selected = torch.multinomial(
                probabilities,
                num_samples=1
            )

            # Get actual vocabulary index
            next_token = top_k_indices.gather(
                -1,
                selected
            )

            # Save token
            generated.append(next_token.item())

            # Append token for next prediction
            input_ids = torch.cat(
                [input_ids, next_token],
                dim=1
            )

    return decode(generated)
```
