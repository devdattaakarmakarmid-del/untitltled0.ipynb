```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import requests
```

```python
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"

text = requests.get(url).text

print("Total Characters :", len(text))
print()
print(text[:500])
```

```python
characters = sorted(list(set(text)))

vocab_size = len(characters)

print("Vocabulary Size :", vocab_size)

# Character to Integer
char_to_int = {}

# Integer to Character
int_to_char = {}

for i, ch in enumerate(characters):
    char_to_int[ch] = i
    int_to_char[i] = ch
```

```python
def encode(sentence):

    numbers = []

    for ch in sentence:
        numbers.append(char_to_int[ch])

    return numbers


def decode(numbers):

    sentence = ""

    for num in numbers:
        sentence += int_to_char[num]

    return sentence
```

```python
sample = "Hello"

encoded = encode(sample)

decoded = decode(encoded)

print("Original :", sample)
print("Encoded :", encoded)
print("Decoded :", decoded)
```

```python
data = torch.tensor(encode(text), dtype=torch.long)

print(data.shape)

print(data[:20])
```

```python
class ShakespeareDataset(Dataset):

    def __init__(self, data, sequence_length):

        self.data = data
        self.sequence_length = sequence_length

    def __len__(self):

        return len(self.data) - self.sequence_length

    def __getitem__(self, index):

        x = self.data[index:index+self.sequence_length]

        y = self.data[index+1:index+self.sequence_length+1]

        return x, y
```

```python
sequence_length = 64

dataset = ShakespeareDataset(data, sequence_length)

print("Dataset Size :", len(dataset))
```

```python
batch_size = 32

loader = DataLoader(
    dataset,
    batch_size=batch_size,
    shuffle=True
)
```

```python
# Model Hyperparameters

d_model = 128          # Embedding size
max_seq_length = 64    # Same as sequence length
```

```python
# Convert token IDs into vectors

token_embedding = nn.Embedding(
    num_embeddings=vocab_size,
    embedding_dim=d_model
)

print(token_embedding)
```

```python
# Learn the position of each token

position_embedding = nn.Embedding(
    num_embeddings=max_seq_length,
    embedding_dim=d_model
)

print(position_embedding)
```

```python
token_vectors = token_embedding(x)

print(token_vectors.shape)
```

```python
position_vectors = position_vectors.unsqueeze(0)

position_vectors = position_vectors.expand(batch_size, -1, -1)

print(position_vectors.shape)
```

```python
embeddings = token_vectors + position_vectors

print(embeddings.shape)
```

```python
import math

class MultiHeadAttention(nn.Module):

    def __init__(self, d_model, num_heads):

        super().__init__()

        self.d_model = d_model
        self.num_heads = num_heads
        self.head_dim = d_model // num_heads

        # Q, K, V Linear Layers
        self.query = nn.Linear(d_model, d_model)
        self.key = nn.Linear(d_model, d_model)
        self.value = nn.Linear(d_model, d_model)

        # Output Layer
        self.output = nn.Linear(d_model, d_model)

    def forward(self, x):

        batch_size = x.size(0)
        seq_len = x.size(1)

        # Create Q, K, V
        Q = self.query(x)
        K = self.key(x)
        V = self.value(x)

        # Split into multiple heads
        Q = Q.view(batch_size, seq_len, self.num_heads, self.head_dim)
        K = K.view(batch_size, seq_len, self.num_heads, self.head_dim)
        V = V.view(batch_size, seq_len, self.num_heads, self.head_dim)

        # Move head dimension
        Q = Q.transpose(1,2)
        K = K.transpose(1,2)
        V = V.transpose(1,2)

        # Attention Scores
        scores = torch.matmul(Q, K.transpose(-2,-1))

        scores = scores / math.sqrt(self.head_dim)

        # Causal Mask
        mask = torch.triu(
            torch.ones(seq_len, seq_len),
            diagonal=1
        )

        mask = mask.bool()

        scores = scores.masked_fill(mask, float('-inf'))

        # Softmax
        attention = torch.softmax(scores, dim=-1)

        # Attention Output
        output = torch.matmul(attention, V)

        # Merge Heads
        output = output.transpose(1,2)

        output = output.contiguous()

        output = output.view(
            batch_size,
            seq_len,
            self.d_model
        )

        output = self.output(output)

        return output
```

```python
num_heads = 8

attention = MultiHeadAttention(
    d_model=d_model,
    num_heads=num_heads
)
```

```python
class FeedForward(nn.Module):

    def __init__(self, d_model):

        super().__init__()

        self.network = nn.Sequential(

            nn.Linear(d_model, d_model * 4),

            nn.GELU(),

            nn.Linear(d_model * 4, d_model)

        )

    def forward(self, x):

        return self.network(x)
```

```python
ffn = FeedForward(d_model)

ffn_output = ffn(attention_output)

print("FFN Output Shape :", ffn_output.shape)
```

```python
class TransformerBlock(nn.Module):

    def __init__(self, d_model, num_heads):

        super().__init__()

        self.attention = MultiHeadAttention(
            d_model,
            num_heads
        )

        self.ffn = FeedForward(d_model)

        self.norm1 = nn.LayerNorm(d_model)

        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):

        # Multi-Head Attention
        attention_output = self.attention(self.norm1(x))

        # Residual Connection
        x = x + attention_output

        # Feed Forward Network
        ffn_output = self.ffn(self.norm2(x))

        # Residual Connection
        x = x + ffn_output

        return x
```

```python
transformer = TransformerBlock(
    d_model=d_model,
    num_heads=num_heads
)

print(transformer)
```

```python
class SimpleTransformer(nn.Module):

    def __init__(self,
                 vocab_size,
                 d_model,
                 num_heads,
                 max_seq_length):

        super().__init__()

        self.token_embedding = nn.Embedding(vocab_size, d_model)

        self.position_embedding = nn.Embedding(max_seq_length, d_model)

        self.transformer = TransformerBlock(
            d_model,
            num_heads
        )

        self.output_layer = nn.Linear(
            d_model,
            vocab_size
        )

    def forward(self, x):

        batch_size = x.size(0)
        seq_len = x.size(1)

        positions = torch.arange(seq_len, device=x.device)
        positions = positions.unsqueeze(0).expand(batch_size, seq_len)

        token = self.token_embedding(x)
        position = self.position_embedding(positions)

        x = token + position

        x = self.transformer(x)

        logits = self.output_layer(x)

        return logits
```

```python
model = SimpleTransformer(
    vocab_size=vocab_size,
    d_model=128,
    num_heads=8,
    max_seq_length=64
)

print(model)
```

```python
loss_function = nn.CrossEntropyLoss()

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=3e-4
)
```

```python
x, y = next(iter(loader))

prediction = model(x)

print(prediction.shape)
```

```python
loss = loss_function(

    prediction.view(-1, vocab_size),

    y.view(-1)

)

print(loss.item())
```

```python
# Number of training epochs
epochs = 1

# Select CPU or GPU automatically
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move model to device
model = model.to(device)

print("Training on:", device)
print("-" * 40)

# Start training
model.train()

for epoch in range(epochs):

    total_loss = 0

    for batch_index, (x, y) in enumerate(loader):

        # Move batch to device
        x = x.to(device)
        y = y.to(device)

        # Clear old gradients
        optimizer.zero_grad()

        # Forward pass
        prediction = model(x)

        # Calculate loss
        loss = loss_function(
            prediction.view(-1, vocab_size),
            y.view(-1)
        )

        # Backpropagation
        loss.backward()

        # Update weights
        optimizer.step()

        # Add batch loss
        total_loss += loss.item()

        # Print every 50 batches
        if (batch_index + 1) % 50 == 0:

            print(
                f"Epoch {epoch+1} | "
                f"Batch {batch_index+1}/{len(loader)} | "
                f"Loss = {loss.item():.4f}"
            )

    average_loss = total_loss / len(loader)

    print("\n-----------------------------------")
    print(f"Epoch {epoch+1} Completed")
    print(f"Average Loss : {average_loss:.4f}")
    print("-----------------------------------\n")

print("Training Finished Successfully!")
```

```python
model.eval()

text = "ROMEO"

encoded = encode(text)

input_ids = torch.tensor(
    encoded,
    dtype=torch.long
).unsqueeze(0)

generated = encoded.copy()

for i in range(100):

    current = input_ids[:, -64:]

    with torch.no_grad():

        output = model(current)

    next_token = output[:, -1, :].argmax(dim=-1)

    generated.append(next_token.item())

    input_ids = torch.cat(
        [input_ids, next_token.unsqueeze(0)],
        dim=1
    )

print(decode(generated))
```
